### ฟังก์ชัน

```python
def CreatesuggestionList(datf) : 
  x = datf['ItemProductRef'].value_counts()
  Suggest_Item_List.append(x.index.values[0])
  Suggest_Item_List.append(x.index.values[1])
  Suggest_Item_List.append(x.index.values[2])
```

```python
def CreatesuggestionListCus(datf) :
  x = datf['ItemProductRef'].value_counts()
  if len(x.index.values) < 3 :
    for i in x.index.values :
      Suggest_Item_List.append(i)
  else :
    Suggest_Item_List.append(x.index.values[0])
    Suggest_Item_List.append(x.index.values[1])
    Suggest_Item_List.append(x.index.values[2])
```

```python
x = '350d25cbd408ef2439302e452e8e7c405e081166'
Suggest_Item_List = []
if x not in tchdat2['CustomerId'].unique() :
  CreatesuggestionList(tchdat2)
  print(1)
elif x in tchdat2['CustomerId'].unique() :
  Cuhdat = tchdat2.loc[tchdat2['CustomerId'] == x]
  CreatesuggestionListCus(Cuhdat)
  print(2)
```

```python
header = ['CustomerId','ItemProductRef']
fin = open('/content/SuggestItem.csv','w',encoding = 'utf-8')
fin = csv.writer(fin)
fin.writerow(header)
for i in cdat['CustomerId'] :
  x = i
  Suggest_Item_List = []
  if x not in tchdat2['CustomerId'].unique() :
    CreatesuggestionList(tchdat2)
  elif x in tchdat2['CustomerId'].unique() :
    Cuhdat = tchdat2.loc[tchdat2['CustomerId'] == x]
    CreatesuggestionListCus(Cuhdat)
  for j in Suggest_Item_List : 
      rw = [i,j]
      fin.writerow(rw)
```
